```python
import re

def convert_storage(storage):
    # Find all numbers followed by 'GB' or 'TB'
    matches = re.findall(r'(\d+)(GB|TB)', storage)

    total_storage = 0
    for value, unit in matches:
        value = int(value)
        if unit == 'TB':  # Convert TB to GB
            value *= 1024
        total_storage += value  # Sum up storage

    return total_storage

df['Memory'] = df['Memory'].apply(convert_storage)


```
